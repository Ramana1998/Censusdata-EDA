```python
var = 'Gender'
prob = data[var].value_counts(normalize = True).reset_index()
plt.subplot(1,1,1)
sns.barplot(x = 'index', y = var, data = prob)
plt.xlabel(var)
plt.ylabel('index')
plt.title('data')
plt.show()
```

```python
var = 'Race'
prob = data[var].value_counts(normalize = True).reset_index()
plt.subplot(1,1,1)
sns.barplot(x = 'index', y = var, data = prob)
plt.xlabel(var)
plt.ylabel('index')
plt.title('Race')
plt.show()
```

```python
var = 'Income'
prob = data[var].value_counts(normalize = False).reset_index()
plt.subplot(1,1,1)
sns.barplot(x = 'index', y = var, data = prob)
plt.xlabel(var)
plt.ylabel('Index')
plt.title('data')
plt.show()
```

```python
var = 'Income'
p = data[var].value_counts(normalize= True).reset_index()
sns.barplot(x = 'index',y = var , data = p)
plt.xlabel(var)
plt.ylabel('Index')
plt.title('Income')
plt.show()
```

```python
var = 'Relationship'
prob = data[var].value_counts(normalize = False).reset_index()
plt.subplot(1,1,1)
sns.barplot(x = 'index', y = var, data = prob)
plt.xlabel(var)
plt.ylabel('Index')
plt.title('Relationship')
plt.show()
```
